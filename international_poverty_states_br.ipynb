```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
#Carregando nosso DF
international_poverty_states_br = pd.read_csv('taxa_pobreza_internacional_estados.csv', sep =';')
```

```python
international_poverty_states_br.head()
```

```python
#Selecionando apenas as colunas que vamos precisar
international_poverty_states_br = international_poverty_states_br.drop(['Sigla', 'Estado'], axis = 1)
```

```python
international_poverty_states_br.head()
```

```python
#Transpondo nosso DF
international_poverty_states_br_T = international_poverty_states_br.T
international_poverty_states_br_T = international_poverty_states_br_T.set_index(international_poverty_states_br.columns)
```

```python
print(international_poverty_states_br_T)
```

```python
#Criando o index superior
international_poverty_states_br_T = international_poverty_states_br_T.reset_index(drop=False)
```

```python
print(international_poverty_states_br_T)
```

```python
print(international_poverty_states_br_T.iloc[0])
```

```python
#Removendo o index superior e renomeando as colunas
new_columns_names = international_poverty_states_br_T.iloc[0].tolist()
```

```python
new_columns_names = dict(zip(international_poverty_states_br_T.columns, new_columns_names))
international_poverty_states_br_T.rename(columns = new_columns_names, inplace=True)
```

```python
print(international_poverty_states_br_T)
```

```python
#Removendo a primeira linha repetida
international_poverty_states_br_T = international_poverty_states_br_T.drop(index=0)
print(international_poverty_states_br_T)
```

```python
international_poverty_states_br_T.head()
```

```python
#Renomeando a variável Código para time
international_poverty_states_br_T.rename(columns = {'Código' : 'time'}, inplace = True)
print(international_poverty_states_br_T)
```

```python
international_poverty_states_br_T = international_poverty_states_br_T.replace(',', '.', regex=True)
print(international_poverty_states_br_T)
```

```python
international_poverty_states_br_T[[11,12,13,14,15,16,17,21,22,23,24,25,26,27,28,29,31,32,33,35,41,42,43,50,51,52,53]] = international_poverty_states_br_T[[11,12,13,14,15,16,17,21,22,23,24,25,26,27,28,29,31,32,33,35,41,42,43,50,51,52,53]].astype(float)
print(international_poverty_states_br_T)
```

```python
international_poverty_states_br_T['time'] = international_poverty_states_br_T['time'].astype(int)
print(international_poverty_states_br_T)
```

```python
#Salvando nosso DF
international_poverty_states_br = international_poverty_states_br_T
international_poverty_states_br.to_csv('international_poverty_states_br.csv',encoding='UTF-8', index = False)
```
