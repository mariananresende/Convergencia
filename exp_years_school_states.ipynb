```python
exp_years_school_states = pd.read_excel('D:\docs_mariana\ProjetosGit\Convergencia\data\Expectativa de estudo.xlsx')
```

```python
exp_years_school_states.head()
```

```python
# Filtrar linhas onde a coluna 'Territorialidades' contenha 'Brasil'
exp_years_school_states = exp_years_school_states[~exp_years_school_states['Territorialidades'].str.contains('Brasil')]
```

```python
exp_years_school_states.head()
```

```python
exp_years_school_states.columns
```

```python
exp_years_school_states = exp_years_school_states.rename(columns={'Expectativa de anos de estudo aos 18 anos de idade 1991': '1991',
       'Expectativa de anos de estudo aos 18 anos de idade 2000': '2000',
       'Expectativa de anos de estudo aos 18 anos de idade 2010': '2010'})
```

```python
exp_years_school_states.head()
```

```python
exp_years_school_states.tail()
```

```python
exp_years_school_states = exp_years_school_states.melt(id_vars=['Territorialidades'], var_name='Ano', value_name='Expect_est')
exp_years_school_states.head()
```

```python
exp_years_school_states.info()
```

```python
exp_years_school_states = exp_years_school_states.dropna()
```

```python
exp_years_school_states.head(28)
```

```python
# Às vezes, o pandas pode interpretar espaços em branco como valores não nulos. 
# Combinação de métodos para garantir que qualquer valor de espaço em branco seja tratado como nulo. 
exp_years_school_states['Territorialidades'] = exp_years_school_states['Territorialidades'].replace(r'^\s*$', np.nan, regex=True)
```

```python
exp_years_school_states = exp_years_school_states.drop(exp_years_school_states[exp_years_school_states['Territorialidades'].isnull()].index)
```

```python
exp_years_school_states.head(28)
```

```python
exp_years_school_states['time'] = exp_years_school_states['Ano'].astype(int)
```

```python
exp_years_school_states.info()
```

```python
exp_years_school_states = exp_years_school_states.drop(columns=['Ano'])
```

```python
exp_years_school_states.head()
```

```python
exp_years_school_states['Expect_est'] = exp_years_school_states['Expect_est'].astype(float)
```

```python
exp_years_school_states.info()
```

```python
exp_years_school_states.head()
```

```python
# Reorganizando o DataFrame para ter os estados como colunas e datas como índices
exp_years_school_states = exp_years_school_states.pivot(index='time', columns='Territorialidades')['Expect_est']

# Removendo o nome da coluna de estados ('Territorialidades')
exp_years_school_states.columns.name = None
```

```python
exp_years_school_states.head()
```

```python
exp_years_school_states.reset_index(inplace=True)
```

```python
exp_years_school_states.head()
```

```python
exp_years_school_states.rename(columns = {'Acre' : '12', 'Alagoas' : '27', 'Amazonas' : '13', 'Amapá' : '16', 'Bahia' : '29', 'Ceará' : '23', 'Distrito Federal' : '53', 'Espírito Santo' : '32', 'Goiás' : '52', 'Maranhão' : '21', "Minas Gerais" : '31', 'Mato Grosso do Sul' : '50', 'Mato Grosso' : '51', 'Pará' : '15', 'Paraíba' : '25', 'Pernambuco' : '26', 'Piauí' : '22', 'Paraná' : '41', 'Rio de Janeiro' : '33', 'Rio Grande do Norte' : '24', 'Rondônia' : '11', 'Roraima' : '14', 'Rio Grande do Sul' : '43', 'Santa Catarina' : '42', 'Sergipe' : '28', 'São Paulo' : '35', 'Tocantins' : '17'}, inplace = True)
exp_years_school_states.head()
```
